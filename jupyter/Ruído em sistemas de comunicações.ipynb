# Ruído em sistemas de comunicações


## Ruído passa-faixa

A recepção de uma portadora modulada, cujo espectro encontra-se centrado numa frequência $f_c$, requer a filtragem passa-faixa dos componentes de frequência da banda $B$ ocupada pelo sinal modulado. Assumindo que o sinal modulado está corrompido por um ruído aditivo gaussiano branco e um processo de filtragem passa-banda ideal, os componentes de ruído fora do intervalo $f_c-B/2<f<f_c+B/2$ serão eliminados e a densidade espectral de potência do ruído exibirá a característica passa-faixa ilustrada na figura abaixo

<img src="./figuras/BandPassNoise.png" width="500">.

Matematicamente, o ruído passa-faixa $n(t)$, ou ruído banda-estreita, pode ser representado por

$$
\begin{equation}
n(t)= n_I(t) \cos \left(2 \pi f_{c} t\right) - n_Q(t) \operatorname{sen}\left(2 \pi f_{c} t\right),
\end{equation}
$$

em que $n_I(t)$ e $n_Q(t)$ representam duas componentes de ruído em banda base, ou passa-baixa, também conhecidas como componentes *em fase* e *quadratura*, respectivamente. 

Após a filtragem passa-faixa, apenas as componentes de ruído dentro do intervalo $f_c-B/2<f<f_c+B/2$ terão influência no processo de demodulação que ocorre no receptor. Desse modo, para obtermos parâmetros que caracterizarão o impacto do ruído
no processo de demodulação, primeiramente se faz necessário entender as propriedades do ruído passa-faixa $n(t)$.

### Caracterização do ruído passa-faixa

**Teorema**: seja $n(t)$ um processo estacionário no sentido amplo (WSS) cuja densidade espectral de potência $S_n(f)$ é nula se $f$ é tal que $f_c-B/2<f<f_c+B/2$, as seguintes relações são válidas

1. $n_I(t)$ e $n_Q(t)$ são processos WSS de média nula.<br><br>

2. As funções de autocorrelação de $n_I(t)$ e $n_Q(t)$ são idênticas, ou seja, $R_{n_I}(\tau)=R_{n_Q}(\tau)$.<br><br>

3. As funções de correlação cruzada de $n_I(t)$ e $n_Q(t)$ são idênticas em valor absoluto, diferindo apenas por um sinal negativo, ou seja, $R_{n_In_Q}(\tau)=-R_{n_Qn_I}(\tau)$.<br><br>

4. $R_n(\tau)  = R_{n_I}(\tau)\cos \left(2 \pi f_{c} \tau\right)+R_{n_Qn_I}(\tau)\operatorname{sen} \left(2 \pi f_{c} \tau\right)$.<br><br>

5. $S_n(f)=\frac{1}{2}\left[S_{n_I}\left(f-f_c\right)+S_{n_I}\left(f+f_c\right)\right]+\frac{1}{2 j}\left[S_{n_Q n_I}\left(f-f_c\right)-S_{n_Q n_I}\left(f+f_c\right)\right].$<br><br>

### Autocorrelação e densidade espectral de potência

O sinal $n(t)$ é um processo estacionário com função de autocorrelação dada por

$$\begin{equation} 
R_{n}(t+\tau, t) = E\left[n(t)n(t+\tau)\right] 
\end{equation}
$$

Podemos expandir o produto $n(t)n(t+\tau)$ na forma

$$
\begin{align} 
n(t)n(t+\tau) &= n_I(t)n_I(t+\tau) \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)+\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]
                 - n_I(t)n_Q(t+\tau) \frac{1}{2}\left[\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\\
                 &- n_Q(t)n_I(t+\tau) \frac{1}{2}\left[-\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen}\left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]
                 + n_Q(t)n_Q(t+\tau) \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)-\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\\
\end{align}
$$

Desse modo, aplicando o valor esperado

$$
\begin{align} 
E\left[n(t)n(t+\tau)\right]  &= E\left[n_I(t)n_I(t+\tau)\right] \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)+\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
                 &- E\left[n_I(t)n_Q(t+\tau)\right] \frac{1}{2}\left[\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
                 &- E\left[n_Q(t)n_I(t+\tau)\right] \frac{1}{2}\left[-\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
                 &+ E\left[n_Q(t)n_Q(t+\tau)\right] \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)-\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
\end{align}
$$

Desse modo, temos que

$$
\begin{align} 
R_n(\tau)  &= \frac{R_{n_I}(\tau)}{2}\cos \left(2 \pi f_{c} \tau\right)+\frac{R_{n_I}(\tau)}{2}\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\nonumber\\
                 &- \frac{R_{n_In_Q}(\tau)}{2}\operatorname{sen} \left(2 \pi f_{c} \tau\right)-\frac{R_{n_In_Q}(\tau)}{2}\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\nonumber\\
                 &+\frac{R_{n_Qn_I}(\tau)}{2}\operatorname{sen} \left(2 \pi f_{c} \tau\right)-\frac{R_{n_Qn_I}(\tau)}{2}\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\nonumber\\
                 &+ \frac{R_{n_Q}(\tau)}{2}\cos \left(2 \pi f_{c} \tau\right)-\frac{R_{n_Q}(\tau)}{2}\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\nonumber\\
\end{align}
$$
Reagrupando os termos, obtemos

$$
\begin{align} 
R_n(\tau)  &= \left[\frac{R_{n_I}(\tau)+R_{n_Q}(\tau)}{2}\right]\cos \left(2 \pi f_{c} \tau\right)+\left[\frac{R_{n_I}(\tau)-R_{n_Q}(\tau)}{2}\right]\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\nonumber\\
                 &+\left[\frac{R_{n_Qn_I}(\tau)-R_{n_In_Q}(\tau)}{2}\right]\operatorname{sen} \left(2 \pi f_{c} \tau\right)-\left[\frac{R_{n_In_Q}(\tau)+R_{n_Qn_I}(\tau)}{2}\right]\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\nonumber\\   \end{align}
$$

Uma vez que $n(t)$ é um processo estacionário, $R_n(\tau)$ deve depender apenas de $\tau$ e não de $t$. Da expressão obtida acima, isto implica que $R_{n_I}(\tau)-R_{n_Q}(\tau)=0$ e $R_{n_In_Q}(\tau)+R_{n_Qn_I}(\tau)=0$, ou seja, $R_{n_I}(\tau)=R_{n_Q}(\tau)$ e $R_{n_In_Q}(\tau)=-R_{n_Qn_I}(\tau)$. 

Finalmente, temos que

$$
\begin{equation} 
R_n(\tau)  = R_{n_I}(\tau)\cos \left(2 \pi f_{c} \tau\right)+R_{n_Qn_I}(\tau)\operatorname{sen} \left(2 \pi f_{c} \tau\right).\end{equation}
$$


Logo, pelo teorema de Wiener-Khinchin, a densidade espectral de potência $S_{n}(f)$ de $n(t)$ será dada por

$$
\begin{align}
S_{n}(f) &= \mathscr{F}[R_n(\tau)]\nonumber\\
&= \mathscr{F}\left[R_{n_I}(\tau)\cos \left(2 \pi f_{c} \tau\right)+R_{n_Qn_I}(\tau)\operatorname{sen} \left(2 \pi f_{c} \tau\right)\right] \nonumber\\
&= \mathscr{F}\left[R_{n_I}(\tau)\right]\ast \mathscr{F}\left[\cos(2\pi f_c \tau)\right] + \mathscr{F}\left[R_{n_Qn_I}(\tau)\right]\ast \mathscr{F}\left[\operatorname{sen}(2\pi f_c \tau)\right] \nonumber\\
&= S_{n_I}(f)\ast \frac{1}{2}\left[\delta(f-f_c)+\delta(f+f_c)\right] + S_{n_Qn_I}(f)\ast \frac{1}{2j}\left[\delta(f-f_c)-\delta(f+f_c)\right]. \nonumber\\
&= \frac{1}{2}\left[S_{n_I}(f-f_c)+S_{n_I}(f+f_c)\right] + \frac{1}{2j}\left[S_{n_Qn_I}(f-f_c)-S_{n_Qn_I}(f+f_c)\right]. \nonumber
\end{align}
$$

Observações:

1. Se os processos $n_I(t)$ e $n_Q(t)$ forem descorrelacionados (o que implica indepedência no caso gaussiano), i.e. se $R_{n_In_Q}(\tau)=0$, temos
$$ S_{n}(f) = \frac{1}{2}\left[S_{n_I}(f-f_c)+S_{n_I}(f+f_c)\right] $$
Nesse caso, uma vez que $R_{n_I}(\tau)$ é uma função real e par, $S_{n_I}(f)$ é real e simétrica com relação a $f=0$ e $S_{n}(f)$ será simétrica em torno de $f=f_c$.<br>


2. Se $R_{n_In_Q}(\tau)\neq 0$, $S_{n_Qn_I}(f)$ deve ser uma função puramente imaginária, uma vez que $S_{n}(f)$ continua sendo real e simétrica em torno de $f=0$, porém assimétrica em torno de $f=f_c$.

### Obtendo $R_{n_I}(\tau)$, $R_{n_Q}(\tau)$ e $R_{n_In_Q}(\tau)$ a partir de $R_{n}(\tau)$

$$
\begin{equation}
\left\{\begin{array}{l}
R_{n_{I}}(\tau)=R_{n_{Q}}(\tau)=R_{n}(\tau) \cos \left(2 \pi f_{0} \tau\right)+\hat{R}_{n}(\tau) \sin \left(2 \pi f_{0} \tau\right) \\
R_{n_{I} n_{Q}}(\tau)=R_{n}(\tau) \sin \left(2 \pi f_{0} \tau\right)-\hat{R}_{n}(\tau) \cos \left(2 \pi f_{0} \tau\right)
\end{array}\right.
\end{equation}
$$

### Obtendo ${S}_{n_I}(f)$, ${S}_{n_Q}(f)$  e $S_{n_I n_Q}(f)$ a partir de $S_{n}(f)$

$$
\begin{equation}
S_{n_I}(f)=S_{n_Q}(f) = \begin{cases}S_{n}\left(f-f_{c}\right)+S_{n}\left(f+f_{c}\right) & |f|<f_c \\ 0, & \text { caso contrário }\end{cases}
\end{equation}
$$

$$
\begin{equation}
S_{n_I n_Q}(f)= \begin{cases}j\left[S_{n}\left(f+f_c\right)-S_{n}\left(f-f_c\right)\right] & |f|<f_c \\ 0, & \text { caso contrário }\end{cases}
\end{equation}
$$

### AM-DSB-SC

#### Sinal modulado $Y(t)$
$$
\begin{equation}
Y(t) = M(t)A\cos(2\pi f_c t +\phi_c)
\end{equation}
$$

#### Função de autocorrelação

$$
\begin{equation}
\bar{R}_{Y}(\tau) = \frac{A^2}{2}R_M(\tau)\cos(2\pi f_c \tau)\nonumber
\end{equation}
$$

#### Ruído passa-faixa

Considerando que um ruído passa-faixa $n(t)$ é adicionado ao sinal transmitido, o sinal $r(t)$ entregue ao demodulador será dado por

$$
\begin{align}
r(t) &= Y(t) + n(t)\nonumber \\
&=A M(t) \cos \left(2 \pi f_{c} t+\phi_{c}\right)+n_I(t) \cos (2 \pi f_{c} t)-n_Q(t) \operatorname{sen}(2 \pi f_{c} t)\nonumber
\end{align}
$$

#### SNR pré-demodulação

Potência de sinal antes da demodulação: $$ P_s =\bar{R}_{Y}(0) = \frac{A^2}{2}P_m $$
Potência de ruído antes da demodulação (ruído passa-faixa): $$ P_n = \int_{-\infty}^{\infty}S_{n}(f)df = 2.\frac{N_0}{2}.2B = 2N_0B$$

Desse modo

$$ 
\begin{align}
 \mathrm{SNR}_{pre}  &= \frac{P_s}{P_n} \nonumber\\
                     &= \frac{\frac{A^2P_m}{2}}{2N_0B}. \nonumber\\
\end{align}
$$
Logo, 

$$
\begin{equation}
  \mathrm{SNR}_{pre} = \frac{A^2P_m}{4N_0B}.
 \end{equation}
$$

#### SNR pós-demodulação

$$
\begin{align}
r(t) \cos \left(2 \pi f_{c} t+\phi\right)=& A M(t) \cos \left(2 \pi f_{c} t+\phi_{c}\right) \cos \left(2 \pi f_{c} t+\phi\right)+n(t) \cos \left(2 \pi f_{c} t+\phi\right) \nonumber\\
=& \frac{1}{2} A M(t) \cos \left(\phi_{c}-\phi\right)+\frac{1}{2} A M(t) \cos \left(4 \pi f_{c} t+\phi+\phi_{c}\right)\nonumber \\
&+\frac{1}{2}\left[n_I(t) \cos \phi+n_Q(t) \operatorname{sen} \phi\right] \nonumber\\
&+\frac{1}{2}\left[n_I(t) \cos \left(4 \pi f_{c} t+\phi\right)-n_Q(t) \operatorname{sen} \left(4 \pi f_{c} t+\phi\right)\right]\nonumber
\end{align}
$$

Potência de sinal após a demodulação: $$ P_s = \left(\frac{A}{2}\right)^2P_m \cos^2 \left(\phi_{c}-\phi\right)$$
Potência de ruído após a demodulação (ruído passa-baixa): $$ P_n = \left(\frac{1}{2}\right)^2(P_{n_I}\cos^2\phi + P_{n_Q}\operatorname{sen}^2\phi) $$

Como 

$$ P_{n_I} = P_{n_Q}  = \int_{-B}^{B}S_{n_I}(f)df = \int_{-B}^{B}N_0 df = 2N_0B $$

temos $$ P_n = \frac{2N_0B}{4}$$

Desse modo, assumindo $\phi_{c}=\phi$, temos

$$ 
\begin{align}
 \mathrm{SNR}_{pós}  &= \frac{P_s}{P_n} \nonumber\\
                     &= \frac{\frac{A^2P_m}{4}}{\frac{2N_0B}{4}}. \nonumber\\
\end{align}
$$

Portanto,
$$
\begin{equation}
  \mathrm{SNR}_{pós} = \frac{A^2P_m}{2N_0B} 
 \end{equation}
$$

#### Ganho de demodulação $\eta$

$$
\begin{equation}
   \eta = \frac{\mathrm{SNR}_{pós}}{\mathrm{SNR}_{pre}} = 2
 \end{equation}
$$

$$
\begin{equation}
   \mathrm{SNR}_{pós[dB]} = \mathrm{SNR}_{pre[dB]} + 3\mathrm{dB}
 \end{equation}
$$

### AM-SSB

#### Sinal modulado $Y(t)$

$$
\begin{equation}
Y(t)=\frac{A}{\sqrt{2}} M(t) \cos \left(2 \pi f_{c} t + \phi_c\right) \pm \frac{A}{\sqrt{2}} \hat{M}(t) \operatorname{sen}\left(2 \pi f_{c} t + \phi_c\right),
\end{equation}
$$

em que $\hat{M}(t)$ é a transformada de Hilbert de $M(t)$.

#### Função de autocorrelação

$$
\begin{equation}
\bar{R}_{Y}(\tau) = \frac{A^{2}}{2} R_{M}(\tau)\cos \left(2 \pi f_{c} \tau\right) \pm \frac{A^{2}}{2}R_{M \hat{M}}(\tau) \operatorname{sen}\left(2 \pi f_{c} \tau\right).
\end{equation}
$$

#### Ruído passa-faixa

Considerando que um ruído passa-faixa $n(t)$ é adicionado ao sinal transmitido, o sinal $r(t)$ entregue ao demodulador será dado por

$$
\begin{align}
r(t) &= Y(t) + n(t)\nonumber \\
&=\frac{A}{\sqrt{2}} M(t) \cos \left(2 \pi f_{c} t + \phi_c\right) \pm \frac{A}{\sqrt{2}} \hat{M}(t) \operatorname{sen}\left(2 \pi f_{c} t + \phi_c\right) + n_I(t) \cos (2 \pi f_{c} t)-n_Q(t) \operatorname{sen}(2 \pi f_{c} t)\nonumber
\end{align}
$$

#### SNR pré-demodulação

Potência de sinal antes da demodulação: $$ P_s =\bar{R}_{Y}(0) = \frac{A^2}{2}P_m $$
Potência de ruído antes da demodulação (ruído passa-faixa): $$ P_n = \int_{-\infty}^{\infty}S_{n}(f)df = \frac{N_0}{2}.2B = N_0B$$

Desse modo

$$ 
\begin{align}
 \mathrm{SNR}_{pre}  &= \frac{P_s}{P_n} \nonumber\\
                     &= \frac{\frac{A^2P_m}{2}}{N_0B}. \nonumber\\
\end{align}
$$
Logo, 

$$
\begin{equation}
  \mathrm{SNR}_{pre} = \frac{A^2P_m}{2N_0B}.
 \end{equation}
$$

#### SNR pós-demodulação

$$
\begin{align}
r(t) \cos \left(2 \pi f_{c} t+\phi\right) = & \left[\frac{A}{\sqrt{2}} M(t) \cos \left(2 \pi f_{c} t + \phi_c\right) \pm \frac{A}{\sqrt{2}} \hat{M}(t) \operatorname{sen}\left(2 \pi f_{c} t + \phi_c\right) \right] \cos \left(2 \pi f_{c} t+\phi\right)+n(t) \cos \left(2 \pi f_{c} t+\phi\right) \nonumber\\
&= \frac{A M(t)}{2\sqrt{2}}  \cos \left(\phi_{c}-\phi\right)+ \frac{A M(t)}{2\sqrt{2}} \cos \left(4 \pi f_{c} t+\phi+\phi_{c}\right)\nonumber \\
&\pm\frac{A \hat{M}(t)}{2\sqrt{2}}  \operatorname{sen} \left(\phi_{c}-\phi\right) \mp \frac{A \hat{M}(t)}{2\sqrt{2}} \operatorname{sen} \left(4 \pi f_{c} t+\phi+\phi_{c}\right)\nonumber \\
&+\frac{1}{2}\left[n_I(t) \cos \phi+n_Q(t) \operatorname{sen} \phi\right] \nonumber\\
&+\frac{1}{2}\left[n_I(t) \cos \left(4 \pi f_{c} t+\phi\right)-n_Q(t) \operatorname{sen} \left(4 \pi f_{c} t+\phi\right)\right]\nonumber
\end{align}
$$

Potência de sinal após a demodulação: $$ P_s = \frac{A^2}{8}P_m \cos^2 \left(\phi_{c}-\phi\right)$$
Potência de ruído após a demodulação (ruído passa-baixa): $$ P_n = \left(\frac{1}{2}\right)^2(P_{n_I}\cos^2\phi + P_{n_Q}\operatorname{sen}^2\phi) $$

Como 

$$ P_{n_I} = P_{n_Q}  = \int_{-B}^{B}S_{n_I}(f)df = \int_{-B}^{B}\frac{N_0}{2} df = \frac{N_0}{2}.2B = N_0B $$

temos $$ P_n = \frac{N_0B}{4}$$

Desse modo, assumindo $\phi_{c}=\phi$, temos

$$ 
\begin{align}
 \mathrm{SNR}_{pós}  &= \frac{P_s}{P_n} \nonumber\\
                     &= \frac{\frac{A^2P_m}{8}}{\frac{N_0B}{4}}. \nonumber\\
\end{align}
$$

Portanto,
$$
\begin{equation}
  \mathrm{SNR}_{pós} = \frac{A^2P_m}{2N_0B} 
 \end{equation}
$$

#### Ganho de demodulação $\eta$

$$
\begin{equation}
   \eta = \frac{\mathrm{SNR}_{pós}}{\mathrm{SNR}_{pre}} = 1
 \end{equation}
$$

$$
\begin{equation}
   \mathrm{SNR}_{pós[dB]} = \mathrm{SNR}_{pre[dB]}
 \end{equation}
$$

### QAM

#### Sinal modulado $Y(t)$

$$
\begin{equation}
Y(t) = A M_1(t) \cos \left(2 \pi f_{c} t + \phi_c\right) - A M_2(t) \operatorname{sen}\left(2 \pi f_{c} t + \phi_c\right),
\end{equation}
$$

#### Função de autocorrelação

$$
\begin{equation}
\bar{R}_{Y}(\tau) = \frac{A^{2}}{2} \left[R_{M_1}(\tau) + R_{M_2}(\tau)\right]\cos \left(2 \pi f_{c} \tau\right).
\end{equation}
$$

#### Ruído passa-faixa

Considerando que um ruído passa-faixa $n(t)$ é adicionado ao sinal transmitido, o sinal $r(t)$ entregue ao demodulador será dado por

$$
\begin{align}
r(t) &= Y(t) + n(t)\nonumber \\
&=AM_1(t) \cos \left(2 \pi f_{c} t + \phi_c\right) -AM_2(t) \operatorname{sen}\left(2 \pi f_{c} t + \phi_c\right) + n_I(t) \cos (2 \pi f_{c} t)-n_Q(t) \operatorname{sen}(2 \pi f_{c} t)\nonumber
\end{align}
$$

#### SNR pré-demodulação

Potência de sinal antes da demodulação: $$ P_s =\bar{R}_{Y}(0) = \frac{A^2}{2}\left[P_{m_1} + P_{m_2}\right] $$
Potência de ruído antes da demodulação (ruído passa-faixa): $$ P_n = \int_{-\infty}^{\infty}S_{n}(f)df = 2.\frac{N_0}{2}.2B = 2N_0B$$

Desse modo

$$ 
\begin{align}
 \mathrm{SNR}_{pre}  &= \frac{P_s}{P_n} \nonumber\\
                     &= \frac{\frac{A^2\left[P_{m_1} + P_{m_2}\right]}{2}}{2N_0B}. \nonumber\\
\end{align}
$$
Logo, 

$$
\begin{equation}
  \mathrm{SNR}_{pre} = \frac{A^2\left[P_{m_1} + P_{m_2}\right]}{4N_0B}.
 \end{equation}
$$

#### SNR pós-demodulação

Potência de sinal após a demodulação: $$ P_{s_1} = \left(\frac{A}{2}\right)^2P_{m_1} \cos^2 \left(\phi_{c}-\phi\right) $$ $$P_{s_2} = \left(\frac{A}{2}\right)^2P_{m_2} \cos^2 \left(\phi_{c}-\phi\right)$$

Potência de ruído após a demodulação (ruído passa-baixa): $$ P_{n_1} = \left(\frac{1}{2}\right)^2(P_{n_I}\cos^2\phi + P_{n_Q}\operatorname{sen}^2\phi) $$ $$ P_{n_2} = \left(\frac{1}{2}\right)^2(P_{n_I}\operatorname{sen}^2\phi + P_{n_Q}\cos^2\phi) $$

Como 

$$ P_{n_I} = P_{n_Q}  = \int_{-B}^{B}S_{n_I}(f)df = \int_{-B}^{B}N_0 df = 2N_0B $$

temos $$ P_{n_1} = P_{n_2} = P_n = \frac{2N_0B}{4}$$

Desse modo, assumindo $\phi_{c}=\phi$, temos

$$ 
\begin{align}
 \mathrm{SNR}_{pós(1,2)}  &= \frac{P_{s_{(1,2)}}}{P_n} \nonumber\\
                     &= \frac{\frac{A^2P_{m_{(1,2)}}}{4}}{\frac{2N_0B}{4}}. \nonumber\\
\end{align}
$$

Portanto,
$$
\begin{equation}
  \mathrm{SNR}_{pós(1,2)} = \frac{A^2P_{m_{(1,2)}}}{2N_0B} 
 \end{equation}
$$

#### Ganho de demodulação $\eta$

$$
\begin{equation}
   \eta = \frac{\mathrm{SNR}_{pós}}{\mathrm{SNR}_{pre}} = 1
 \end{equation}
$$

$$
\begin{equation}
   \mathrm{SNR}_{pós[dB]} = \mathrm{SNR}_{pre[dB]}
 \end{equation}
$$

### AM-DSB

#### Sinal modulado $Y(t)$

$$
\begin{equation}
Y(t) = [1 + k_aM(t)]A\cos(2\pi f_c t).
\end{equation}
$$

sendo $k_a$ o índice de modulação. 

#### Função de autocorrelação

$$
\begin{equation}
\bar{R}_{Y}(\tau) = \frac{A^2}{2}\left[1 + 2k_am + k_a^2R_M(\tau)\right]\cos(2\pi f_c \tau)\nonumber
\end{equation}
$$

Considerando que um ruído passa-faixa $n(t)$ é adicionado ao sinal transmitido, o sinal $r(t)$ entregue ao demodulador será dado por

#### Ruído passa-faixa

$$
\begin{align}
r(t) &= Y(t) + n(t)\nonumber \\
&=[1 + k_aM(t)]A\cos(2\pi f_c t) + n_I(t) \cos (2 \pi f_{c} t)-n_Q(t) \operatorname{sen}(2 \pi f_{c} t)\nonumber
\end{align}
$$

#### SNR pré-demodulação

Potência de sinal antes da demodulação: $$ P_s = \frac{k_a^2A^2}{2}P_m $$
Potência de ruído antes da demodulação (ruído passa-faixa): $$ P_n = \int_{-\infty}^{\infty}S_{n}(f)df = 2.\frac{N_0}{2}.2B = 2N_0B$$

Desse modo

$$ 
\begin{align}
 \mathrm{SNR}_{pre}  &= \frac{P_s}{P_n} \nonumber\\
                     &= \frac{\frac{k_a^2A^2P_m}{2}}{2N_0B}. \nonumber\\
\end{align}
$$
Logo, 

$$
\begin{equation}
  \mathrm{SNR}_{pre} = \frac{k_a^2A^2P_m}{4N_0B}.
 \end{equation}
$$

#### SNR pós-demodulação

Assumindo que o sinal $r(t)$ entrege ao detector de envoltória é afetado apenas pela presença do ruído passa-faixa, podemos  escrever

$$
\begin{align}
r(t) &= \left\{A\left[1+k_a M(t)\right] + n_I(t)\right\} \cos(2\pi f_c t) - n_Q(t) \operatorname{sen}(2\pi f_c t)\nonumber\\
r(t) &= r_I(t) \cos(2\pi f_c t) - r_Q(t) \operatorname{sen}(2\pi f_c t)\nonumber\\
r(t) &= E_r(t)\cos\left[2\pi f_c t + \Theta_r(t)\right]\nonumber
\end{align}
$$

em que

$$
\begin{align}
E_r(t)&=\sqrt{r_I^{2}(t)+r_Q^{2}(t)} \nonumber\\
\Theta_r(t)&=\arctan -\frac{r_Q(t)}{r_I(t)}.\nonumber
\end{align}
$$

Logo, a envoltória $E_r(t)$ do sinal $r(t)$ será dada por
$$
\begin{equation}
E_r(t) = \sqrt{\left[A\left[1+k_a M(t)\right]+n_I(t)\right]^{2}+n_Q^{2}(t)}.
\end{equation}
$$

Se o sistema opera num regime de alta $\mathrm{SNR}$, temos que $A\left[1+k_a M(t)\right]\gg n_Q(t)$ e podemos aproximar

$$
\begin{align}
E_r(t)&=\sqrt{\left[A\left[1+k_a M(t)\right]+n_I(t)\right]^{2}+n_Q^{2}(t)}\nonumber\\
      &\approx\sqrt{\left[A\left[1+k_a M(t)\right]+n_I(t)\right]^{2}}\nonumber\\
      &\approx A\left[1+k_a M(t)\right]+n_I(t).\nonumber
\end{align}
$$

Desse modo,

$$ P_n = P_{n_I}  = \int_{-B}^{B}S_{n_I}(f)df = \int_{-B}^{B}N_0 df = 2N_0B $$

$$ 
\begin{align}
 \mathrm{SNR}_{pos}  &= \frac{P_s}{P_n} \nonumber\\
                     &= \frac{k_a^2A^2P_m}{2N_0B}. \nonumber\\
\end{align}
$$

Nesse caso, o ganho de demodulação $\eta$ é o mesmo do caso AM-DSB-SC

$$
\begin{equation}
   \eta = \frac{\mathrm{SNR}_{pós}}{\mathrm{SNR}_{pre}} = 2
 \end{equation}
$$

$$
\begin{equation}
   \mathrm{SNR}_{pós[dB]} = \mathrm{SNR}_{pre[dB]} + 3\mathrm{dB}
 \end{equation}
$$

Portanto, no regime de alta $\mathrm{SNR}$, AM-DSB-SC e AM-DSB com detecção de envoltória são equivalentes. 

Considerando a situação inversa, isto é, que a potência de ruído é muito maior do que a potência de sinal, teremos que

$$
\begin{align}
E_r(t) &=\sqrt{\left[A\left[1+k_a M(t)\right]+n_I(t)\right]^{2}+n_Q^{2}(t)} \\
&=\sqrt{A^{2}\left[1+k_a M(t)\right]^{2}+n_I^{2}(t)+n_Q^{2}(t)+2 A n_I(t)\left[1+k_a M(t)\right]}. \\
\end{align}
$$

Nesse caso, o termo $A^2\left[1+k_a M(t)\right]^2$ será muito menor do que os demais e podemos aproximar

$$
\begin{equation}
E_r(t) \approx \sqrt{\left[n_I^{2}(t)+n_Q^{2}(t)\right]\left[1+\frac{2 A n_I(t)}{n_I^{2}(t)+n_Q^{2}(t)}\left[1+k_a M(t)\right]\right]}\nonumber.
\end{equation}
$$

Seja $E_n(t) =\sqrt{n_I^{2}(t)+n_Q^{2}(t)}$ e $\sqrt{1+\epsilon} \approx 1+\frac{\epsilon}{2}$ para valores pequenos de $\epsilon$, em que $\epsilon=\frac{2 A n_I(t)}{n_I^{2}(t)+n_Q^{2}(t)}\left[1+k_a M(t)\right]$, temos

$$
\begin{align}
E_r(t) & \approx \sqrt{\left[n_I^{2}(t)+n_Q^{2}(t)\right]\left[1+\frac{2 A n_I(t)}{n_I^{2}(t)+n_Q^{2}(t)}\left[1+k_a M(t)\right]\right]}\nonumber \\
& \approx E_n(t)\left[1+\frac{A n_I(t)}{E_n^{2}(t)}\left[1+k_a M(t)\right]\right]\nonumber \\
&=E_n(t)+\frac{A n_I(t)}{E_n(t)}\left[1+k_a M(t)\right]\nonumber
\end{align}
$$

Perceba que nesse caso o ruído não tem mais apenas uma característica aditiva, mas também afeta o sinal por um meio de uma componente multiplicativa. Nessa situação, não é possível definir a $\mathrm{SNR}$.

## Impacto do ruído em modulações de ângulo


Em sistemas que empregam modulação de ângulo, a informação é mapeada na fase ou na frequência da portadora. Nesses sistemas, o sinal mensagem será afetado pelo ruído na medida em que este altera o conteúdo de frequências do sinal modulado.

$$
\begin{array}{rlr}Y(t) & =A \cos \left(2 \pi f_{c} t+ \Phi_M(t)\right) \\ & = \begin{cases}A \cos \left(2 \pi f_{c} t+2 \pi k_{f} \int_{-\infty}^{t} M(\tau) d \tau\right), & \mathrm{FM} \\ A \cos \left(2 \pi f_{c} t+k_{p} M(t)\right), & \mathrm{PM}\end{cases} \end{array}$$

ou seja,

$$
\Phi_M(t)= \begin{cases}k_{p} M(t), & \mathrm{PM} \\ 2 \pi k_{f} \int_{-\infty}^{t} M(\tau) d \tau, & \mathrm{FM}\end{cases}
$$

$$
\begin{align}
r(t) &= Y(t) + n(t)\nonumber \\
&= Y(t) + n_I(t) \cos (2 \pi f_{c} t)-n_Q(t) \operatorname{sen}(2 \pi f_{c} t)\nonumber\\
&= Y(t) + \sqrt{n_I^2(t)+n_Q^2(t)} \cos \left(2 \pi f_{c} t+\arctan \frac{n_Q(t)}{n_I(t)}\right)\nonumber\\
&= A \cos \left(2 \pi f_{c} t+ \Phi_M(t)\right) + V_n(t) \cos \left(2 \pi f_{c} t+\Phi_{n}(t)\right)
\end{align}
$$


em que $V_n(t)$ e $\Phi_{n}(t)$ representam a amplitude e a fase da envoltória do ruído passa-faixa $n(t)$.

<img src="./figuras/FasoresRuidoSinaisPMFM.png?" width="500">

Assumindo que a potência da portadora modulada é muito maior que a potência de ruído, ou seja, que $p\left(V_{n}(t) \ll A\right) \approx 1$, podemos escrever as seguintes aproximações para o sinal recebido

$$
\begin{align}
r(t) &\approx  \left[A+V_{n}(t) \cos \alpha(t)\right] \cos \left(2 \pi f_{c} t+\Phi_M(t)+\arctan \frac{V_{n}(t) \sin \alpha(t)}{A+V_{n}(t) \cos \alpha(t)}\right) \\
& \approx  \left[A + V_{n}(t) \cos \alpha(t) \right] \cos \left(2 \pi f_{c} t+\Phi_M(t)+\frac{V_{n}(t)}{A} \operatorname{sen} \alpha(t)\right)
\end{align}
$$

com $\alpha(t) = \Phi_n(t)-\Phi_M(t)$. No processo de demodulação haverá uma conversão da fase de $r(t)$ para um sinal de amplitude a partir do qual o sinal mensagem deve ser recuperado. Seja $\hat{M}(t)$ o sinal na saída do demodulador, então

$$
\begin{align}
\hat{M}(t) &=\left\{\begin{array}{l}
k_p M(t)+\frac{V_{n}(t)}{A} \operatorname{sen} \alpha(t),\quad \mathrm{PM} \nonumber \\
k_f M(t)+\frac{1}{2\pi}\frac{d}{dt} \left[\frac{V_n(t)}{A} \operatorname{sen} \alpha(t)\right],\quad \mathrm{FM}
\end{array}\right.\\
&=\left\{\begin{array}{l}
k_p M(t)+ N(t),\quad \mathrm{PM} \\
k_f M(t)+\frac{1}{2\pi}\frac{dN(t)}{dt} ,\quad \mathrm{FM}
\end{array}\right.
\end{align}
$$

em que $N(t)=\frac{V_{n}(t)}{A} \operatorname{sen} \alpha(t)$ é o termo de ruído associado à saída do demodulador.

$$
\begin{align} 
N(t) &=\frac{V_{n}(t)}{A} \operatorname{sen} \left(\Phi_{n}(t)-\Phi_M(t)\right)\nonumber\\ 
         &=\frac{1}{A}\left[V_{n}(t) \sin \Phi_{n}(t) \cos \Phi_M(t)-V_{n}(t) \cos \Phi_{n}(t) \operatorname{sen} \Phi_M(t)\right] \nonumber\\ 
         &=\frac{1}{A}\left[n_Q(t) \cos \Phi_M(t)-n_I(t) \operatorname{sen} \Phi_M(t)\right] \nonumber
\end{align}
$$

Considerando que $R_{n_I}(\tau)=R_{n_Q}(\tau)$ e $R_{n_In_Q}(\tau)=0$, a autocorrelação de $N(t)$ será dada por

$$
\begin{align}
E\left[N(t+\tau) N(t)\right]=& \frac{1}{A^{2}} E\left[R_{n_Q}(\tau) \cos \Phi_M(t) \cos \Phi_M(t+\tau)+R_{n_I}(\tau) \operatorname{sen} \Phi_M(t+\tau) \operatorname{sen} \Phi_M(t)\right] \nonumber\\
                                    =& \frac{1}{A^{2}} R_{n_I}(\tau) E\left\{\cos \left[\Phi_M(t+\tau)-\Phi_M(t)\right]\right\}
\end{align}
$$

Se assumirmos que a mensagem $M(t)$ é um processo estacionário gaussiano de média nula e função de autocorrelação denotada por $R_M(\tau)$, também serão gaussianos os processos $\Phi_M(t)$, tanto no caso do sinal PM quanto no caso do sinal FM. Logo, para qualquer instante $t$ a variável $U(t,\tau) = \Phi_M(t+\tau)-\Phi_M(t)$ é a diferença entre duas variáveis aleatórias gaussianas conjuntas. Logo, $U(t,\tau)$ também será uma variável aleatória gaussiana com média nula e variância dada por 

$$
\begin{aligned}
\sigma_{U}^{2} &=E\left[\Phi_M^{2}(t+\tau)\right]+E\left[\Phi_M^{2}(t)\right]-2 R_{\Phi_M}(\tau) \\
&=2\left[R_{\Phi_M}(0)-R_{\Phi_M}(\tau)\right]
\end{aligned}
$$

Portanto,

$$
\begin{align}
E\left[N(t+\tau) N(t)\right] &=\frac{1}{A^{2}} R_{n_I}(\tau) E[\cos (\Phi_M(t+\tau)-\Phi_M(t))] \nonumber\\
&=\frac{1}{A^{2}} R_{n_I}(\tau) \operatorname{Re}\left[E e^{j(\Phi_M(t+\tau)-\Phi_M(t))}\right] \nonumber\\
&=\frac{1}{A^{2}} R_{n_I}(\tau) \operatorname{Re}\left[E e^{j U(t, \tau)}\right] \nonumber\\
&=\frac{1}{A^{2}} R_{n_I}(\tau) \operatorname{Re}\left[e^{-\frac{1}{2} \sigma_{U}^{2}}\right] \nonumber\\
&=\frac{1}{A^{2}} R_{n_I}(\tau) \operatorname{Re}\left[e^{-\left(R_{\Phi_M}(0)-R_{\Phi_M}(\tau)\right)}\right] \nonumber\\
&=\frac{1}{A^{2}} R_{n_I}(\tau) e^{-\left(R_{\Phi_M}(0)-R_{\Phi_M}(\tau)\right)}\nonumber
\end{align}
$$

onde utilizou-se que para uma variavel aletória gaussiana $X\sim\mathcal{N}(\mu,\sigma)$ com função característica $\Psi_{X}(v)$, temos
$$
\begin{align}
\Psi_{X}(v) &=\int_{-\infty}^{\infty}\left[\frac{1}{\sqrt{2 \pi \sigma^{2}}} e^{\frac{-(x-\mu)^{2}}{2 \sigma^{2}}}\right] e^{j v x} d x \nonumber\\
&= E\left[e^{j v X} \right]=e^{j v \mu-\frac{v^{2} \sigma^{2}}{2}}\nonumber
\end{align}
$$

e portanto, para $v=1$ e $\mu=0$, $E\left[e^{jX} \right]=e^{-\frac{\sigma^{2}}{2}}$.

Como $E\left[N(t+\tau) N(t)\right]$ é função apenas de $\tau$, conclui-se que $N(t)$ é um processo estacionário com função de autocorrelação dada por $R_N(\tau)=\frac{1}{A^2} R_{n_I}(\tau) e^{-\left(R_{\Phi_M}(0)-R_{\Phi_M}(\tau)\right)}$. Pelo teorema de Wiener-Kinchin, segue-se que

$$
\begin{align}
S_{N}(f) &=\mathscr{F}\left[R_{N}(\tau)\right]\nonumber \\
&=\mathscr{F}\left[\frac{1}{A^{2}} R_{n_I}(\tau) e^{-\left(R_{\Phi_M}(0)-R_{\Phi_M}(\tau)\right)}\right]\nonumber \\
&=\frac{e^{-R_{\Phi_M}(0)}}{A^{2}} \mathscr{F}\left[R_{n_I}(\tau) e^{R_{\Phi_M}(\tau)}\right] \nonumber\\
&=\frac{e^{-R_{\Phi_M}(0)}}{A^{2}} \mathscr{F}\left[R_{n_I}(\tau) g(\tau)\right]\nonumber \\
&=\frac{e^{-R_{\Phi_M}(0)}}{A^{2}} S_{n_I}(f) \ast G(f)\nonumber
\end{align}
$$

em que $g(\tau) = e^{R_{\Phi_M}(\tau)}$ e $G(f)=\mathscr{F}\left[g(\tau)\right]$.

$$
\begin{align}
S_{N}(f) &=\frac{e^{-R_{\Phi_M}(0)}}{A^{2}} S_{n_I}(f) \ast G(f) \nonumber\\
&=\frac{e^{-R_{\Phi_M}(0)}}{A^{2}} N_{0} \int_{-B_{c}}^{B_{c}} G(f) df \nonumber\\
& \approx \frac{e^{-R_{\Phi_M}(0)}}{A^{2}} N_{0} \int_{-\infty}^{\infty} G(f) df\nonumber \\
&=\left.\frac{e^{-R_{\Phi_M}(0)}}{A^{2}} N_{0} g(\tau)\right|_{\tau=0}\nonumber\\
&=\frac{e^{-R_{\Phi_M}(0)}}{A^{2}} N_{0} e^{R_{\Phi_M}(0)} \nonumber\\
&=\frac{N_{0}}{A^{2}}\nonumber\\
\end{align}
$$

Portanto, a densidade espectral de potência de $N(t)$ será dada por

$$
S_{N}(f)= \begin{cases}\frac{N_{0}}{A^{2}}, & \mathrm{PM} \\ \frac{N_{0}}{A^{2}} f^{2}, & \mathrm{FM}\end{cases}.
$$

Com $S_{N}(f)$ podemos calcular a potência de ruído na saída do demodulador no caso do sinal PM e do sinal FM como
$$
\begin{align}
P_{N} &=\int_{-B}^{B} S_{N}(f) df \nonumber\\
&= \begin{cases}\int_{-B}^{B} \frac{N_{0}}{A^{2}} df, & \text { PM } \\ \int_{-B}^{B} \frac{N_{0}f^2}{A^{2}} df, & \text { FM }\end{cases}\nonumber\\
&= \begin{cases}\frac{2N_{0}B}{A^{2}}, & \mathrm{PM} \\ \frac{2 N_{0} B^{3}}{3 A^{2}}, & \mathrm{FM.}\end{cases}\nonumber
\end{align}
$$

A potência de sinal na saída do demodulador, por sua vez, será dada por
$$
P_s = \begin{cases}k_{p}^{2} P_{M}, & \text { PM } \\ k_{f}^{2} P_{M}, & \text { FM. }\end{cases}
$$

Por fim, obtemos a $\mathrm{SNR}$ do sinal demodulado

$$\large
\mathrm{SNR} = \begin{cases}\frac{k_{p}^{2} A^{2} P_{M}}{2N_{0}B}, & \text { PM } \\ \frac{3k_{f}^{2} A^{2} P_{M}}{2N_{0}B^3}, & \text { FM. }\end{cases}
$$


$$
\begin{cases}\beta_{p}=k_{p} \max |M(t)|, & \mathrm{PM} \\ \beta_{f}=\frac{k_{f} \max |M(t)|}{B}, & \mathrm{FM}\end{cases}
$$
