```python
def genGIF(x, y, figName, xlabel=[], ylabel=[], fram=200, inter=20):    
    '''
    Create and save a plot animation as GIF
     
    :param x: x-axis values [np array]
    :param y: y-axis values [np array]
    :param figName: figure file name w/ folder path [string]
    :param xlabel: xlabel [string]
    :param ylabel: ylabel [string]
    :param fram: number of frames [int]
    :param inter: time interval between frames [milliseconds]
    
    '''
    figAnin = plt.figure()
    ax      = plt.axes(xlim=(np.min(x)-0.2, np.max(x)+0.2),\
                       ylim=(np.min(y)-0.2*np.max(np.abs(y)), np.max(y)+0.2*np.max(np.abs(y))))
    line,   = ax.plot([], [], 'o', markersize = 15)
    plt.axis('square')
    ax.grid()  

    indx = np.arange(0, len(x), int(len(x)/fram))
    
    if len(xlabel): 
           plt.xlabel(xlabel)
            
    if len(ylabel): 
           plt.ylabel(ylabel)
        
    def init():        
        line.set_data([], [])
        return line,

    def animate(i):
        line.set_data(x[indx[i]], y[indx[i]])
        
        if x[indx[i]]>0 and y[indx[i]]>0:
            line.set_color("blue")
        elif x[indx[i]]>0 and y[indx[i]]<0:
            line.set_color("red")
        elif x[indx[i]]<0 and y[indx[i]]<0:
            line.set_color("green")
        else:
            line.set_color("orange")

        return line,

    anim = FuncAnimation(figAnin, animate, init_func=init, frames=fram, interval=inter, blit=True)

    anim.save(figName, dpi=200, writer='imagemagick')
    plt.close()
```

```python
def cond(A, B):    
    return np.logical_and(A,B)
```

### Definindo um exemplo de experimento aleatório

```python
eventos = np.array([[1,1],[-1,1],[-1,-1],[1,-1]])

for ind in range(0,4):
    print('Evento %d: o ponto mostrado no gráfico é (%d,%d)'%(ind, eventos[ind,0], eventos[ind,1]))
```

### Atribuindo probabilidades a eventos

```python
prob = np.array([[3/10, 2/10, 4/10, 1/10]]).T
#prob = np.array([[1/4, 1/4, 1/4, 1/4]]).T

for ind in range(0,4):
    print('P(Evento %d) = %.2f'%(ind, prob[ind,0]))
```

```python
for ind in range(0,4):
    x = eventos[ind,0]
    y = eventos[ind,1]
    lab = round(prob[ind,0],2)
    plt.plot(x, y,'o', markersize = 15,label=lab)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    
plt.axis('square')
plt.grid()
```

```python
# gera uma sequência de realizações de acordo com a probabilidade associada a cada evento
samples = np.random.choice(np.arange(4), 100, p=prob[:,0]) 

x = eventos[samples,0]
y = eventos[samples,1]

figName = "H:\\Meu Drive\\DEE UFCG\\Disciplinas e ementas\\Sistemas de Comunicações\\códigos\\jupyter\\sim1.gif"
genGIF(x, y, figName, xlabel='x', ylabel='y', fram=len(samples), inter=1000)

Image('sim1.gif', width=500)
```

#### Definindo novos eventos

Evento A : $x>0$

Evento B : $x<0$

Evento C : $y>0$

Evento D : $y<0$

#### Calculando a probabilidade P(A, C)

```python
x = eventos[:,0]
y = eventos[:,1]

# calculando a probabilidade de (A,C) 
pAC = sum(prob[cond(x>0, y>0)])

print('P(A,C) = ', round(pAC[0],2))
```

#### Calculando as probabilidades P(A) e P(B)

```python
# calculando a probabilidade do evento A e do evento B
pA = sum(prob[x>0])
pB = sum(prob[x<0])

print('P(A) = ', round(pA[0],2))
print('P(B) = ', round(pB[0],2))
```

#### Calculando as probabilidades de P(C) e P(D)

```python
# calculando a probabilidade do evento C e do evento D 
pC = sum(prob[y>0])
pD = sum(prob[y<0])

print('P(C) = ', round(pC[0],2))
print('P(D) = ', round(pD[0],2))
```

#### Calculando as probabilidades condicionais P(A|C) e P(C|A)

```python
# calculando as probabilidades condicionais P(A|C) e P(C|A) 
pAdC = pAC/pC
pCdA = pAC/pA

print('P(A|C) = ', round(pAdC[0],2))
print('P(C|A) = ', round(pCdA[0],2))
```
