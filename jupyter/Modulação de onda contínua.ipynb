# Modulação de onda contínua

## Potência e energia

Considere $X(t)$ um processo estocástico e $x_i(t)$ uma função amostra deste processo. A energia $\mathcal{E}_{i}$ e a potência $P_{i}$ da função $x_i(t)$ serão dadas por

$$\begin{equation}
\begin{gathered}
\mathcal{E}_{i}=\int_{-\infty}^{\infty} x_i^{2}\left(t\right) d t \\
P_{i}=\lim _{T \rightarrow \infty} \frac{1}{T} \int_{-T / 2}^{T / 2} x_i^{2}\left(t\right) d t.
\end{gathered}
\end{equation}$$

Como as funções amostra $x_i(t)$ são aleatórias, energia e potência podem ser definidas como duas variáveis aleatórias reais $\mathscr{E}_{X}$ e $\mathscr{P}_{X}$, respectivamente. A energia e a potência de um processo estocástico podem, então, ser definidas como os valores experados de $\mathscr{E}_{X}$ e $\mathscr{P}_{X}$, ou seja

$$
\begin{align}
P_{X}&=E\left[\mathscr{P}_{X}\right] \\
\mathcal{E}_{X}&=E\left[\mathscr{E}_{X}\right] 
\end{align}
$$

com $\mathscr{E}_{X}=\int_{-\infty}^{\infty} X^{2}(t) d t$ e $\mathscr{P}_{X}=\lim _{T \rightarrow \infty} \frac{1}{T} \int_{-T / 2}^{T / 2} X^{2}(t) d t $. Desse modo, temos que

$$
\begin{align}
\mathcal{E}_{X}&=E\left[\int_{-\infty}^{\infty} X^{2}(t) d t\right] \\
&=\int_{-\infty}^{\infty} E\left[X^{2}(t)\right] d t \\
&=\int_{-\infty}^{\infty} R_{X}(t, t) d t \\
\end{align}
$$

$$
\begin{align}
P_{X}&=E\left[\lim _{T \rightarrow \infty} \frac{1}{T} \int_{-T / 2}^{T / 2} X^{2}(t) d t\right] \\
&=\lim_{T \rightarrow \infty} \frac{1}{T} \int_{-T / 2}^{T / 2} E\left[X^{2}(t)\right] d t \\
&=\lim_{T \rightarrow \infty} \frac{1}{T} \int_{-T / 2}^{T / 2} R_{X}(t, t) d t
\end{align}
$$

Se $X$ for um processo estacionário, temos que $R_{X}(t, t)=R_X(0)$ é independente de $t$, logo

$$
\begin{align}
P_{X}&=R_X(0) \\
\mathcal{E}_{X}&=\int_{-\infty}^{\infty}R_X(0)dt
\end{align}
$$

Observe que se a energia do processo estacionário for finita, i.e. $\mathcal{E}_{X}<\infty$, sua potência será nula, ou seja, $R_X(0)=0$ (processo tipo energia). Isto significa que para todo o $t$, $X(t)$ será zero com probabilidade um. Dessa maneira, apenas os processos estacionários do tipo potência são de interesse teórico e prático.


## Processos cicloestacionários

Um processo estocástico $X(t)$ com média $m_X(t)$ e função de autocorrelação $R_X(t+\tau, t)$ é dito cicloestacionário se tanto a média quanto a função de autocorrelação forem periódicas em $t$ com período $T_0$, ou seja,

$$\begin{equation}
\begin{gathered}
m_{X}\left(t+T_{0}\right)=m_{X}(t) \\
R_{X}\left(t+\tau+T_{0}, t+T_{0}\right)=R_{X}(t+\tau, t)
\end{gathered}
\end{equation}$$

para todo $t$ e todo $\tau$.

Exemplo 1: seja $Y(t) = X(t)\cos(2\pi f_0 t)$, onde $X(t)$ é um processo estacionário de média $m_X$ e autocorrelação $R_X(\tau)$, temos que

$$\begin{align} 
m_Y(t) &= E\left[Y(t)\right]\nonumber \\&= E\left[X(t)\cos(2\pi f_0 t)\right]\nonumber\\ &= E\left[X(t)\right]\cos(2\pi f_0 t)\nonumber \\ &= m_X\cos(2\pi f_0 t) \nonumber
\end{align}$$

$$\begin{align} 
R_Y(t+\tau, t) &= E\left[Y(t+\tau)Y(t)\right]\nonumber \\&= E\left[X(t+\tau)\cos(2\pi f_0 (t+\tau))X(t)\cos(2\pi f_0 t)\right]\nonumber\\ &= E\left[X(t+\tau)X(t)\right]\cos(2\pi f_0 (t+\tau))\cos(2\pi f_0 t)\nonumber \\ &= R_X(\tau)\left[\frac{1}{2}\cos(2\pi f_0 \tau) + \frac{1}{2}\cos(4\pi f_0 t + 2\pi f_0 \tau) \right] \nonumber
\end{align}$$

Logo, tanto $m_Y(t)$ quanto $R_Y(t)$ são periódicas com período $T_0=1/f_0$. Portanto, $Y(t)$ é um processo cicloestacionário.

## Espectro de potência de um processo estocástico


Considere $X(t)$ um processo estocástico e $x_i(t)$ uma função amostra deste processo. Assuma que $x_{i_T}(t)$ é uma versão truncada de $x_i(t)$ no intervalo $-T/2<t<T/2$, ou seja


$$\begin{equation}\label{eq1}
x_{i_T}\left(t \right)= \begin{cases}x_i\left(t \right), & |t|<T/2 \\ 0, & \text { caso contrário. }\end{cases}
\end{equation}
$$

Uma vez que o sinal $x_{i_T}(t)$ tem duração finita, podemos definir sua transformada de Fourier

$$\begin{equation}\label{eq2}
X_{i_T}(f) = \int_{-\infty}^{\infty} x_{i_T}(t)e^{-j2\pi ft}dt
\end{equation}
$$

Sabe-se que a energia $E_T$ do sinal $x_{i_T}(t)$ é definida como

$$\begin{equation}\label{eq3}
E_{i_T} = \int_{-\infty}^{\infty} |x_{i_T}(t)|^2dt = \int_{-\infty}^{\infty} |X_{i_T}(f)|^2df
\end{equation}
$$

em que a segunda igualdade decorre do teorema de Parseval.

Dessa forma, a *densidade espectral de energia* de $x_{i_T}(t)$ é dada por $|X_{i_T}(f)|^2$. Logo, podemos definir *densidade espectral de potência* de $x_{i_T}(t)$ como sendo a *densidade espectral de energia* por unidade de tempo, ou seja, $\frac{|X_{i_T}(f)|^2}{T}$. 

Finalmente, podemos definir a densidade espectral de potência $S_{x_i}(f)$ da função amostra $x_{i_T}(t)$ como

$$\begin{equation}\label{eq4}
S_{x_i}(f) = \lim_{T \to \infty} \frac{|X_{i_T}(f)|^2}{T}.
\end{equation}
$$

Por sua vez, cada função amostra $x_i(t)$ do processo $X(t)$ pode ter uma densidade espectral de potência diferente, de modo que a densidade espectral de potência de $X(t)$ será definida por

$$\begin{equation}\label{eq5}
S_{X}(f) = E\left[\lim_{T \to \infty} \frac{|X_{i_T}(f)|^2}{T}\right] = \lim_{T \to \infty} \frac{E\left[|X_{i_T}(f)|^2\right]}{T}.
\end{equation}
$$

## Teorema de Wiener-Khinchin

Se para todo intervalo finito $\tau$ e qualquer intervalo $\mathcal{A}$ de comprimento $|\tau|$, a função de autocorrelação do processo estocástico $X(t)$ satisfaz a condição


$$ \left| \int_{\mathcal{A}} R_X(t+\tau, t)dt \right|<\infty,$$

então a densidade espectral de potência $S_X(f)$ de $X(t)$ é a transformada de Fourier de $\bar{R}_X(t+\tau, t)$ em que

$$\begin{equation}\label{eq6}
\bar{R}_X(t+\tau, t) = \lim_{T \to \infty} \frac{1}{T}\int_{-T/2}^{T/2}R_X(t+\tau, t)dt.
\end{equation}
$$

Portanto, o teorema estabelece que

$$\begin{equation}\label{eq7}
S_x(f) = \mathscr{F}\left[\lim_{T \to \infty} \frac{1}{T}\int_{-T/2}^{T/2}R_X(t+\tau, t)dt\right].
\end{equation}
$$

**Prova**: Fazendo uso da definição da densidade espectral de potência $S_{X}(f)$ de um processo $X(t)$, temos

$$
\begin{equation}\label{eq8}
S_{X}(f)=\lim _{T \rightarrow \infty} \frac{1}{T} E\left[\left|X_{T}(f)\right|^{2}\right],
\end{equation}
$$

em que 

$$
\begin{equation}\label{eq9}
X_{T}(f)=\int_{-T / 2}^{T / 2} X(t) e^{-j 2 \pi f t} d t.
\end{equation}
$$

De (\ref{eq8}) e (\ref{eq9}), podemos escrever

$$
\begin{aligned}
S_{X}(f) &=\lim _{T \rightarrow \infty} \frac{1}{T} E\left[\int_{-T / 2}^{T / 2} X_{T}(s) e^{-j 2 \pi f s} d s \int_{-T / 2}^{T / 2} X_{T}(t) e^{+j 2 \pi f t} d t\right] \\
&=\lim _{T \rightarrow \infty} \frac{1}{T} \int_{-T / 2}^{T / 2} \int_{-T / 2}^{T / 2} R_{X}(s, t) e^{-j 2 \pi f(s-t)} d t d s.
\end{aligned}
$$

Aplicando a transformada inversa de Fourier, temos

$$
\begin{aligned}
\mathscr{F}^{-1}\left[S_{X}(f)\right] &=\lim _{T \rightarrow \infty} \frac{1}{T} \int_{-\infty}^{\infty} e^{+j 2 \pi f \tau} \int_{-T / 2}^{T / 2} \int_{-T / 2}^{T / 2} R_{X}(s, t) e^{-j 2 \pi f(s-t)} d t d s d f \\
&=\lim _{T \rightarrow \infty} \frac{1}{T} \int_{-T / 2}^{T / 2} \int_{-T / 2}^{T / 2} R_{X}(s, t) d t d s \int_{-\infty}^{\infty} e^{j 2 \pi f[\tau-(s-t)]} d f\\
&=\lim _{T \rightarrow \infty} \frac{1}{T} \int_{-T / 2}^{T / 2} \int_{-T / 2}^{T / 2} R_{X}(s, t)\delta(\tau-s+t) d t d s\\
&=\lim _{T \rightarrow \infty} \frac{1}{T} \int_{-T / 2}^{T / 2} d t \int_{-T / 2}^{T / 2} R_{X}(s, t) \delta(\tau-s+t) d s.
\end{aligned}
$$

Da propriedade da filtragem da função impulso $\int_{-\infty}^{\infty}f(t)\delta(t-a)dt=f(a)$, temos que

$$ \begin{equation}\label{eq10}
\int_{-T / 2}^{T / 2} R_{X}(s, t) \delta(\tau-s+t) d s= \begin{cases}R_{X}(t+\tau, t), & -T / 2<t+\tau<T / 2 \\ 0, & \text { caso contrário. }\end{cases}
\end{equation}
$$

Logo, temos que

$$
\mathscr{F}^{-1}\left[\mathcal{S}_{X}(f)\right]=\lim _{T \to \infty} \frac{1}{T} \int_{-T / 2}^{T / 2}\left\{\begin{array}{ll}
R_{X}(t+\tau, t), & -T / 2<t+\tau<T / 2 \\
0, & \text { caso contrário }
\end{array}\right\} d t.
$$

Perceba que, se $\tau>0$, os limites da integral serão $(-T/2, T/2-\tau)$, enquanto que, se $\tau<0$, os limites da integral serão $(-T/2-\tau, T/2)$. Desse modo, temos que


$$
\mathscr{F}^{-1}\left[\mathcal{S}_{X}(f)\right]= \begin{cases}\lim _{T \to \infty} \frac{1}{T} \int_{-T / 2}^{T / 2-\tau} R_{X}(t+\tau, t) d t, & \tau>0 \\ \lim _{T \to \infty} \frac{1}{T} \int_{-T / 2-\tau}^{T / 2} R_{X}(t+\tau, t) d t, & \tau<0\end{cases}
$$

$$
\mathscr{F}^{-1}\left[\mathcal{S}_{X}(f)\right]=\left\{\begin{array}{l}
\lim _{T \to \infty} \frac{1}{T}\left[\int_{-T / 2}^{T / 2} R_{X}(t+\tau, t) d t-\int_{T / 2-\tau}^{T / 2} R_{X}(t+\tau, t) d t\right], \quad \tau>0 \\
\lim _{T \to \infty} \frac{1}{T}\left[\int_{-T / 2}^{T / 2} R_{X}(t+\tau, t) d t-\int_{-T / 2}^{-T / 2-\tau} R_{X}(t+\tau, t) d t\right], \quad \tau<0
\end{array}\right.
$$

$$
\mathscr{F}^{-1}\left[\mathcal{S}_{X}(f)\right]=\left\{\begin{array}{l}
\lim _{T \to \infty} \frac{1}{T}\left[\int_{-T / 2}^{T / 2} R_{X}(t+\tau, t) d t\right]-\lim _{T \to \infty} \frac{1}{T}\left[\int_{T / 2-\tau}^{T / 2} R_{X}(t+\tau, t) d t\right], \quad \tau>0 \\
\lim _{T \to \infty} \frac{1}{T}\left[\int_{-T / 2}^{T / 2} R_{X}(t+\tau, t) d t\right]-\lim _{T \to \infty} \frac{1}{T}\left[\int_{-T / 2}^{-T / 2-\tau} R_{X}(t+\tau, t) d t\right], \quad \tau<0
\end{array}\right.
$$

Como para todo intervalo finito $\tau$ e qualquer intervalo $\mathcal{A}$ de comprimento $|\tau|$,$ \left| \int_{\mathcal{A}} R_X(t+\tau, t)dt \right|<\infty$, os dois limites do lado direito das expressões serão nulos, e finalmente temos que

$$\begin{equation}\label{eq11}
\mathscr{F}^{-1}\left\lbrace S_x(f)\right\rbrace = \lim_{T \to \infty} \frac{1}{T}\int_{-T/2}^{T/2}R_X(t+\tau, t)dt.
\end{equation}
$$
ou seja,

$$\begin{equation}\label{eq12}
S_x(f) = \mathscr{F}\left[\lim_{T \to \infty} \frac{1}{T}\int_{-T/2}^{T/2}R_X(t+\tau, t)dt\right].
\end{equation}
$$

concluindo a demonstração.

### Corolário 1

Se $X(t)$ for um processo estacionário para o qual $\tau R_X(\tau)$ permanece finito para todo $\tau$ finito, então

$$\begin{equation}\label{eq13}
S_x(f) = \mathscr{F}\left[R_X(\tau)\right].
\end{equation}
$$

### Corolário 2

Caso $X(t)$ seja um processo cicloestacionário, se 

$$
\left|\int_{0}^{T_{0}} R_{X}(t+\tau, t) d t\right|<\infty
$$

a condição do teorema de Wiener-Khinchin é satisfeita e a densidade espectral de potência é obtida por 

$$
\mathcal{S}_{X}(f)=\mathscr{F}\left[\bar{R}_{X}(\tau)\right],
$$

com $\bar{R}_{X}(\tau)$ definida por

$$
\bar{R}_{X}(\tau) \stackrel{\text { def }}{=} \frac{1}{T_{0}} \int_{-T_{0} / 2}^{T_{0} / 2} R_{X}(t+\tau, t) d t,
$$

em que $T_0$ é o período do processo cicloestacionário.

## Sinal AM-DSB-SC

O diagrama de blocos de um transmissor AM-DSB com portadora suprimida está ilustrado na figura a seguir

<img src="./figuras/AMDSBSC.png" width="400">

Assuma que o sinal mensagem que desejamos transmitir por sistema AM-DSB-SC seja um processo $M(t)$ estacionário, com média $m$ e função de autocorrelação $R_M(\tau)$ e densidade espectral de potência $S_M(f)=\mathscr{F}[R_M(\tau)]$. Seja a portadora representada pela função $c(t)=A\cos(2\pi f_c t)$, em que $A$ é amplitude e $f_c$ é a frequência da portadora, podemos representar o sinal modulado pelo processo $Y(t)$, dado por

$$
\begin{equation}
Y(t) = M(t)c(t) = M(t)A\cos(2\pi f_c t).
\end{equation}
$$

### Autocorrelação e densidade espectral de potência

Como vimos no Exemplo 1, $Y(t)$ é um processo cicloestacionário com função de autocorrelação dada por

$$\begin{equation} 
R_{Y}(t+\tau, t) =R_M(\tau)\left[\frac{A^2}{2}\cos(2\pi f_c \tau) + \frac{A^2}{2}\cos(4\pi f_c t + 2\pi f_c \tau) \right] 
\end{equation}$$

Pelo corolário 2 do teorema de Wiener-Khinchin, temos

$$
\begin{align}
\bar{R}_{Y}(\tau) &= \frac{1}{T_{c}} \int_{-T_{c} / 2}^{T_{c} / 2} R_{Y}(t+\tau, t) dt\nonumber\\
&= \frac{A^2}{2}R_M(\tau)\cos(2\pi f_c \tau)\nonumber
\end{align}
$$

em que $T_c = 1/f_c$. Logo, a densidade espectral de potência $S_{Y}(f)$ do sinal modulado será dada por

$$
\begin{align}
S_{Y}(f) &= \mathscr{F}[\bar{R}_{Y}(\tau)]\nonumber\\
&= \mathscr{F}\left[\frac{A^2}{2}R_M(\tau)\cos(2\pi f_c \tau)\right] \nonumber\\
&= \frac{A^2}{2}\mathscr{F}\left[R_M(\tau)\right]\ast \mathscr{F}\left[\cos(2\pi f_c \tau)\right] \nonumber\\
&= \frac{A^2}{2}S_M(f)\ast \frac{1}{2}\left[\delta(f+fc) + \delta(f-fc)\right] \nonumber \\
&= \frac{A^2}{4}\left[S_M(f+fc) + S_M(f-fc)\right] \nonumber
\end{align}
$$

### Potência do sinal mensagem e do sinal modulado

Seja $P_M$ a potência do sinal mensagem, temos que 

$$ P_M = R_M(0) = \int_{-\infty}^{\infty} S_M(f)df $$.

Seja $P_Y$ a potência do sinal modulado, segue que 

$$ P_Y = \bar{R}_Y(0) = \int_{-\infty}^{\infty} S_Y(f)df = \frac{A^2}{2}P_M$$.

## Sinal AM-DSB

O diagrama de blocos de um transmissor AM-DSB com portadora está ilustrado na figura a seguir

<img src="./figuras/AMDSB.png" width="500">

Para o sinal AM-DSB com portadora, podemos representar o sinal modulado por

$$
\begin{equation}
Y(t) = [1 + k_aM(t)]A\cos(2\pi f_c t) = M'(t)A\cos(2\pi f_c t).
\end{equation}
$$

em que $M'(t) = 1 + k_aM(t)$, sendo $k_a$ o índice de modulação. 


### Autocorrelação e densidade espectral de potência

Do caso AM-DSB-SC, sabemos que a seguinte relação é válida

$$
\begin{equation}
S_{Y}(f) = \frac{A^2}{4}\left[S_{M'}(f+fc) + S_{M'}(f-fc)\right] 
\end{equation}
$$

Logo, para determinar $S_{Y}(f)$ é necessário encontrar $S_{M'}(f)$. Temos que 

$$
\begin{align}
R_{M'}(\tau) &= E\left[M'(t+\tau)M'(t)\right]\nonumber\\
&= E\left[(1 + k_aM(t+\tau))(1 + k_aM(t))\right]\nonumber\\
&= E\left[1 + k_aM(t+\tau) + k_aM(t) + k_a^2M(t+\tau)M(t)\right]\nonumber\\
&= 1 + k_aE\left[M(t+\tau)\right] + k_aE\left[M(t)\right] + k_a^2E\left[M(t+\tau)M(t)\right]\nonumber\\
&= 1 + 2k_am + k_a^2R_M(\tau)\nonumber\\
\end{align}
$$

Como $S_{M'}(f) = \mathscr{F}[R_{M'}(\tau)]$, temos que

$$
\begin{align}
S_{M'}(f) &= \mathscr{F}[1 + 2k_am + k_a^2R_M(\tau)]\nonumber\\
&= (1 + 2k_am)\delta(f) + k_a^2S_M(f)\nonumber\\
\end{align}
$$

Finalmente, temos
$$
\begin{align}
S_{Y}(f) &= \frac{A^2}{4}\left[S_{M'}(f+fc) + S_{M'}(f-fc)\right] \nonumber\\
&= \frac{A^2}{4}\left[(1 + 2k_am)\delta(f+fc) + k_a^2S_M(f+fc) + (1 + 2k_am)\delta(f-fc) + k_a^2S_M(f-fc)\right]\nonumber\\
&= \frac{A^2}{4}(1 + 2k_am)\left[\delta(f+fc) + \delta(f-fc)\right] + \frac{A^2k_a^2}{4}\left[ S_M(f+fc) + S_M(f-fc)\right] \nonumber\\
\end{align}
$$


### Potência do sinal modulado

Seja $P_Y$ a potência do sinal modulado, segue que 

$$ P_Y = \bar{R}_Y(0) = \int_{-\infty}^{\infty} S_Y(f)df = \frac{A^2}{2}(1 + 2k_am) + \frac{A^2k_a^2}{2}P_M$$.

## Sinal AM-SSB

O diagrama de blocos de um transmissor AM-SSB está ilustrado na figura a seguir

<img src="./figuras/AMSSB.png" width="800">.

O sinal AM-SSB é representado por

$$
\begin{equation}
Y(t)=\frac{A}{\sqrt{2}} M(t) \cos \left(2 \pi f_{c} t\right) \pm \frac{A}{\sqrt{2}} \hat{M}(t) \operatorname{sen}\left(2 \pi f_{c} t\right),
\end{equation}
$$

em que $\hat{M}(t)$ é a transformada de Hilbert de $M(t)$.

### O filtro de Hilbert

A transformada de hilbert $\mathcal{H}\{x(t)\}$ de um sinal $x(t)$ pode ser entendida como a convolução de $x(t)$ com um filtro cuja resposta ao impulso $h(t)$ é dada por

$$
\begin{equation}
h(t)= \frac{1}{\pi t}
\end{equation}
$$

cuja transformada de Fourier é dada por

$$
\begin{equation}
H(f)= \begin{cases}-j, & f>0 \\ j, & f<0 \\ 0, & f=0\end{cases}.
\end{equation}
$$

Observa-se que resposta em frequência $H(f)$ pode ser escrita em termos da função degrau unitário $u(f)$ como $H(f)=j[u(-f)-u(f)]$.

Assuma que $\hat{M}(t)$ é o resultado da filtragem do processo estacionário $M(t)$ por $h(t)$, ou seja, 

$$ 
\begin{equation}
\hat{M}(t) = M(t)\ast h(t)
\end{equation}
$$

Das propriedades relacionadas à filtragem linear de processos estacionários, temos as seguintes relações entre as densidades espectrais de potência e funções de autocorrelação da entrada e da saída para o filtro de Hilbert:

$$ \begin{array}{ll}S_{\hat{M}}(f)&=|H(f)|^{2} S_{M}(f) & \Rightarrow S_{\hat{M}}(f)=S_{M}(f) \Rightarrow R_{\hat{M}}(\tau)=R_{M}(\tau) \\ S_{\hat{M} M}(f)&=H(f) S_{M}(f) & \Rightarrow S_{\hat{M} M}(f)=j[u(-f)-u(f)] S_{M}(f) \\ S_{M \hat{M}}(f)&=H^{*}(f) S_{M}(f) & \Rightarrow S_{M \hat{M}}(f)=-j[u(-f)-u(f)] S_{M}(f)\end{array}
$$

Perceba que

$$
\begin{equation}
S_{M \hat{M}}(f)=-S_{\hat{M} M}(f) \Rightarrow R_{M \hat{M}}(\tau)=-R_{\hat{M} M}(\tau)
\end{equation}
$$


### Autocorrelação e densidade espectral de potência

Assim como no caso AM-DSB-SC, $Y(t)$ é um processo cicloestacionário definido por

$$ Y(t)=\frac{A}{\sqrt{2}} M(t) \cos \left(2 \pi f_{c} t\right) \pm \frac{A}{\sqrt{2}} \hat{M}(t) \operatorname{sen}\left(2 \pi f_{c} t\right), $$

com função de autocorrelação dada por

$$\begin{equation} 
R_{Y}(t+\tau, t) = E\left[Y(t)Y(t+\tau)\right] 
\end{equation}
$$

Podemos expandir o produto $Y(t)Y(t+\tau)$ na forma

$\begin{align} 
Y(t)Y(t+\tau) &= \frac{A^{2}}{2} M(t)M(t+\tau) \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)+\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]
                 \pm \frac{A^{2}}{2} M(t)\hat{M}(t+\tau) \frac{1}{2}\left[\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\\
                 &\pm \frac{A^{2}}{2} \hat{M}(t)M(t+\tau) \frac{1}{2}\left[-\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen}\left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]
                 + \frac{A^{2}}{2} \hat{M}(t)\hat{M}(t+\tau) \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)-\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\\
\end{align}$

Desse modo, aplicando o valor esperado

$
\begin{align} 
E\left[Y(t)Y(t+\tau)\right]  &= \frac{A^{2}}{2} E\left[M(t)M(t+\tau)\right] \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)+\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]
                 \pm \frac{A^{2}}{2} E\left[M(t)\hat{M}(t+\tau)\right] \frac{1}{2}\left[\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
                 &\pm \frac{A^{2}}{2} E\left[\hat{M}(t)M(t+\tau)\right] \frac{1}{2}\left[-\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]
                 + \frac{A^{2}}{2} E\left[\hat{M}(t)\hat{M}(t+\tau)\right] \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)-\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
\end{align}
$

Pelo corolário 2 do teorema de Wiener-Khinchin, temos

$$
\begin{align}
\bar{R}_{Y}(\tau) &= \frac{1}{T_{c}} \int_{-T_{c} / 2}^{T_{c} / 2} R_{Y}(t+\tau, t) dt\nonumber\\
&= \frac{A^2}{4}R_M(\tau)\cos(2\pi f_c \tau) \pm \frac{A^2}{4}R_{\hat{M}M}(\tau)\operatorname{sen}(2\pi f_c \tau) \\&\mp \frac{A^2}{4}R_{M\hat{M}}(\tau)\operatorname{sen}(2\pi f_c \tau) + \frac{A^2}{4}R_{\hat{M}}(\tau)\cos(2\pi f_c \tau)\nonumber
\end{align}
$$

em que $T_c = 1/f_c$. 

Das propriedades do filtro de Hilbert, sabemos que $R_{M}(\tau)=R_{\hat{M}}(\tau)$ e $R_{M \hat{M}}(\tau)=-R_{\hat{M} M}(\tau)$. Desse modo, a expressão de $\bar{R}_{Y}(\tau)$ pode ser simplificada para

$$
\begin{equation}
\bar{R}_{Y}(\tau) = \frac{A^{2}}{2} R_{M}(\tau)\cos \left(2 \pi f_{c} \tau\right) \pm \frac{A^{2}}{2}R_{\hat{M} M}(\tau) \operatorname{sen}\left(2 \pi f_{c} \tau\right).
\end{equation}
$$

Logo, a densidade espectral de potência $S_{Y}(f)$ do sinal modulado será dada por

$$
\begin{align}
S_{Y}(f) &= \mathscr{F}[\bar{R}_{Y}(\tau)]\nonumber\\
&= \mathscr{F}\left[\frac{A^{2}}{2} R_{M}(\tau)\cos \left(2 \pi f_{c} \tau\right) \pm \frac{A^{2}}{2}R_{\hat{M}M}(\tau) \operatorname{sen}\left(2 \pi f_{c} \tau\right)\right] \nonumber\\
&= \frac{A^2}{2}\mathscr{F}\left[R_M(\tau)\right]\ast \mathscr{F}\left[\cos(2\pi f_c \tau)\right] \pm \frac{A^2}{2}\mathscr{F}\left[R_{\hat{M}M}(\tau)\right]\ast \mathscr{F}\left[\operatorname{sen}(2\pi f_c \tau)\right] \nonumber\\
&= \frac{A^2}{2}S_M(f)\ast \frac{1}{2}\left[\delta(f-f_c)+\delta(f+f_c)\right] \pm \frac{A^2}{2}S_{\hat{M} M}(f)\ast \frac{1}{2j}\left[\delta(f-f_c)-\delta(f+f_c)\right]. \nonumber\\
&= \frac{A^2}{4}\left[S_M(f-f_c)+S_M(f+f_c)\right] \pm \frac{A^2}{4j}\left[S_{\hat{M} M}(f-f_c)-S_{\hat{M} M}(f+f_c)\right]. \nonumber
\end{align}
$$

Desejamos representar a densidade espectral de potência do sinal AM-SSB modulado $S_{Y}(f)$ apenas em função da densidade espectral do sinal mensagem $S_{M}(f)$. Para tanto, utilizamos as seguintes relações obtidas na análise do filtro de Hilbert

$$
\begin{aligned} 
&S_{\hat{M} M}(f) = j[u(-f)-u(f)] S_{M}(f) \\ 
&S_{\hat{M} M}\left(f-f_{c}\right) = j\left[u(-f+f_c)-u\left(f-f_{c}\right)\right] S_{M}\left(f-f_{c}\right) \\ 
&S_{\hat{M} M}\left(f+f_{c}\right) = j\left[u(-f-f_c)-u\left(f+f_{c}\right)\right] S_{M}\left(f+f_{c}\right) 
\end{aligned}
$$

Por conveniência, podemos escrever ainda

$$
\begin{aligned}
&S_{M}(f)=[u(f)+u(-f)] S_{M}(f)\\
&S_{M}\left(f-f_{c}\right)=\left[u\left(f-f_{c}\right)+u\left(-f+f_{c}\right)\right] S_{M}\left(f-f_{c}\right) \\
&S_{M}\left(f+f_{c}\right)=\left[u\left(f+f_{c}\right)+u\left(-f-f_{c}\right)\right] S_{M}\left(f+f_{c}\right)
\end{aligned}
$$

#### AM - SSB - USB

Assumindo o caso da transformada de Hilbert com sinal negativo (SSB com banda lateral superior), temos que

$$
\begin{align}
S_{Y}(f) &= \frac{A^2}{4}\left[S_M(f-f_c)+S_M(f+f_c)\right] - \frac{A^2}{4j}\left[S_{\hat{M} M}(f-f_c)-S_{\hat{M} M}(f+f_c)\right]. \nonumber \\
&= \frac{A^2}{4}\left[S_{M}\left(f-f_{c}\right) +S_{M}\left(f+f_{c}\right)\right] \nonumber \\
&- \frac{A^2}{4}\left[u(-f+f_c)S_{M}\left(f-f_{c}\right)-u\left(f-f_{c}\right)S_{M}\left(f-f_{c}\right)-u(-f-f_c)S_{M}\left(f+f_{c}\right)+u\left(f+f_{c}\right)S_{M}\left(f+f_{c}\right)\right]. \nonumber \\
&= \frac{A^2}{4}S_{M}\left(f-f_{c}\right)\left[1 - u\left(-f+f_{c}\right) + u\left(f-f_{c}\right) \right] +\frac{A^2}{4} S_{M}\left(f+f_{c}\right)\left[1 + u\left(-f-f_{c}\right) - u\left(f+f_{c}\right) \right]. \nonumber \\
&= \frac{A^2}{2}\left[u\left(f-f_{c}\right)S_{M}\left(f-f_{c}\right) +u\left(-f-f_{c}\right)S_{M}\left(f+f_{c}\right)\right]. \nonumber \\
\end{align}
$$

#### AM - SSB - LSB

Assumindo o caso da transformada de Hilbert com sinal positivo (SSB com banda lateral inferior), temos que

$$
\begin{align}
S_{Y}(f) &= \frac{A^2}{4}\left[S_M(f-f_c)+S_M(f+f_c)\right] + \frac{A^2}{4j}\left[S_{\hat{M} M}(f-f_c)-S_{\hat{M} M}(f+f_c)\right]. \nonumber \\
&= \frac{A^2}{4}\left[S_{M}\left(f-f_{c}\right) +S_{M}\left(f+f_{c}\right)\right] \nonumber \\
&+ \frac{A^2}{4}\left[u(-f+f_c)S_{M}\left(f-f_{c}\right)-u\left(f-f_{c}\right)S_{M}\left(f-f_{c}\right)-u(-f-f_c)S_{M}\left(f+f_{c}\right)+u\left(f+f_{c}\right)S_{M}\left(f+f_{c}\right)\right]. \nonumber \\
&= \frac{A^2}{4}S_{M}\left(f-f_{c}\right)\left[1 + u\left(-f+f_{c}\right) - u\left(f-f_{c}\right) \right] +\frac{A^2}{4} S_{M}\left(f+f_{c}\right)\left[1 - u\left(-f-f_{c}\right) + u\left(f+f_{c}\right) \right]. \nonumber \\
&= \frac{A^2}{2}\left[u\left(-f+f_{c}\right)S_{M}\left(f-f_{c}\right) +u\left(f+f_{c}\right)S_{M}\left(f+f_{c}\right)\right]. \nonumber \\
\end{align}
$$

### Potência do sinal mensagem e do sinal modulado

Seja $P_M$ a potência do sinal mensagem, temos que 

$$ P_M = R_M(0) = \int_{-\infty}^{\infty} S_M(f)df $$.

Seja $P_Y$ a potência do sinal modulado, segue que 

$$ P_Y = \bar{R}_Y(0) = \int_{-\infty}^{\infty} S_Y(f)df = \frac{A^2}{2}\left[\frac{P_M}{2} + \frac{P_M}{2} \right]= \frac{A^2}{2}P_M$$.

## Sinal QAM

O diagrama de blocos de um transmissor QAM está ilustrado na figura a seguir

<img src="./figuras/QAM.png" width="800">.

O sinal QAM é representado por

$$
\begin{equation}
Y(t)= M_1(t) \cos \left(2 \pi f_{c} t\right) - M_2(t) \operatorname{sen}\left(2 \pi f_{c} t\right),
\end{equation}
$$

em que $M_1(t)$ e $M_2(t)$ representam duas mensagens independentes e descorrelacionadas.


### Autocorrelação e densidade espectral de potência

O sinal QAM modulado $Y(t)$ é um processo cicloestacionário definido por

$$ Y(t)= M_1(t) A\cos \left(2 \pi f_{c} t\right) - M_2(t) A \operatorname{sen}\left(2 \pi f_{c} t\right), $$

com função de autocorrelação dada por

$$\begin{equation} 
R_{Y}(t+\tau, t) = E\left[Y(t)Y(t+\tau)\right] 
\end{equation}
$$

Podemos expandir o produto $Y(t)Y(t+\tau)$ na forma

$\begin{align} 
Y(t)Y(t+\tau) &= A^{2} M_1(t)M_1(t+\tau) \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)+\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]
                 - A^{2} M_1(t)M_2(t+\tau) \frac{1}{2}\left[\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\\
                 &- A^{2} M_2(t)M_1(t+\tau) \frac{1}{2}\left[-\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen}\left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]
                 + A^{2} M_2(t)M_2(t+\tau) \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)-\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\\
\end{align}$

Desse modo, aplicando o valor esperado

$
\begin{align} 
E\left[Y(t)Y(t+\tau)\right]  &= A^{2} E\left[M_1(t)M_1(t+\tau)\right] \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)+\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
                 &- A^{2} E\left[M_1(t)M_2(t+\tau)\right] \frac{1}{2}\left[\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
                 &- A^{2} E\left[M_2(t)M_1(t+\tau)\right] \frac{1}{2}\left[-\operatorname{sen} \left(2 \pi f_{c} \tau\right)+\operatorname{sen} \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
                 &+ A^{2} E\left[M_2(t)M_2(t+\tau)\right] \frac{1}{2}\left[\cos \left(2 \pi f_{c} \tau\right)-\cos \left(4 \pi f_{c} t+2 \pi f_{c} \tau\right)\right]\nonumber\\
\end{align}
$

Pelo corolário 2 do teorema de Wiener-Khinchin, temos

$$
\begin{align}
\bar{R}_{Y}(\tau) &= \frac{1}{T_{c}} \int_{-T_{c} / 2}^{T_{c} / 2} R_{Y}(t+\tau, t) dt\nonumber\\
&= \frac{A^2}{2}R_{M_1}(\tau)\cos(2\pi f_c \tau) - \frac{A^2}{2}R_{M_1M_2}(\tau)\operatorname{sen}(2\pi f_c \tau) \\&+ \frac{A^2}{2}R_{M_2M_1}(\tau)\operatorname{sen}(2\pi f_c \tau) + \frac{A^2}{2}R_{M_2}(\tau)\cos(2\pi f_c \tau)\nonumber
\end{align}
$$

em que $T_c = 1/f_c$. 

Como as mensagens $M_1(t)$ e $M_2(t)$ são descorrelacionadas, temos que $R_{M_2M_1}(\tau)=R_{M_1M_2}(\tau)=0$.

Desse modo, a expressão de $\bar{R}_{Y}(\tau)$ pode ser simplificada para

$$
\begin{equation}
\bar{R}_{Y}(\tau) = \frac{A^{2}}{2} R_{M_1}(\tau)\cos \left(2 \pi f_{c} \tau\right) + \frac{A^{2}}{2}R_{M_2}(\tau) \cos\left(2 \pi f_{c} \tau\right).
\end{equation}
$$

Logo, a densidade espectral de potência $S_{Y}(f)$ do sinal modulado será dada por

$$
\begin{align}
S_{Y}(f) &= \mathscr{F}[\bar{R}_{Y}(\tau)]\nonumber\\
&= \mathscr{F}\left[\frac{A^{2}}{2} R_{M_1}(\tau)\cos \left(2 \pi f_{c} \tau\right) + \frac{A^{2}}{2}R_{M_2}(\tau) \cos\left(2 \pi f_{c} \tau\right)\right] \nonumber\\
&= \frac{A^2}{2}\mathscr{F}\left[R_{M_1}(\tau)\right]\ast \mathscr{F}\left[\cos(2\pi f_c \tau)\right] + \frac{A^2}{2}\mathscr{F}\left[R_{M_2}(\tau)\right]\ast \mathscr{F}\left[\cos(2\pi f_c \tau)\right] \nonumber\\
&= \frac{A^2}{2}S_{M_1}(f)\ast \frac{1}{2}\left[\delta(f-f_c)+\delta(f+f_c)\right] + \frac{A^2}{2}S_{M_2}(f)\ast \frac{1}{2}\left[\delta(f-f_c)+\delta(f+f_c)\right]. \nonumber\\
&= \frac{A^2}{4}\left[S_{M_1}(f-f_c)+S_{M_1}(f+f_c)\right] + \frac{A^2}{4}\left[S_{M_2}(f-f_c)+S_{M_2}(f+f_c)\right]. \nonumber
\end{align}
$$

### Potência do sinal mensagem e do sinal modulado

Seja $P_M$ a potência do sinal mensagem, temos que 

$$ 
\begin{align}
P_M &= P_{M_1} + P_{M_2}\nonumber\\ 
    &= R_{M_1}(0) + R_{M_2}(0)\nonumber\\
    &= \int_{-\infty}^{\infty} S_{M_1}(f)df + \int_{-\infty}^{\infty} S_{M_2}(f)df
\end{align}
$$.

Seja $P_Y$ a potência do sinal modulado, segue que 

$$ P_Y = \bar{R}_Y(0) = \int_{-\infty}^{\infty} S_Y(f)df = \frac{A^2}{4}\left[2P_{M_1} + 2P_{M_2} \right]= \frac{A^2}{2}P_M$$.

## Modulação em ângulo

No caso dos esquemas de modulação em fase (*phase modulation*-PM) ou em frequência (*frequency modulation*-FM), o sinal modulado $Y(t)$ pode ser expresso como

$$
\begin{equation}
Y(t) = A\cos\left[2\pi f_ct + \Theta(t)\right]
\end{equation}
$$

sendo $\Theta(t)$ é uma função do sinal mensagem $M(t)$ dada por

$$
\begin{equation}
\Theta(t)= \begin{cases}k_{p} M(t), & \mathrm{PM} \\ 2 \pi k_{f} \int_{-\infty}^{t} M(\tau) d \tau, & \mathrm{FM,}\end{cases}
\end{equation}
$$

em que $k_p$ e $k_f$ são as constantes de proporcionalidade de desvio de fase e frequência, respectivamente.

Uma vez que a fase é o parâmetro modulado, o espectro do sinal modulado dependerá do quanto a modulação expande os componentes de frequência da portadora. Para um sinal PM, definimos o desvio máximo de fase $\Delta \Theta_{\mathrm{PM}}$ e o desvio máximo de frequência  $\Delta f_{\mathrm{PM}}$ por

$$
\begin{aligned}
&\Delta \Theta_{\mathrm{PM}}=k_{p} \max [|M(t)|]\\
&\Delta f_{\mathrm{PM}}= \frac{1}{2\pi} \frac{d\Theta(t)}{dt} = \frac{k_p}{2\pi}  \max \left[\left|\frac{dM(t)}{dt}\right|\right]\\
\end{aligned}
$$

No caso da modulação sinal FM, temos

$$
\begin{aligned}
&\Delta \Theta_{\mathrm{FM}}=2\pi k_f \max \left[\left| \int_{-\infty}^{t} M(\tau) d \tau\right|\right]\\
&\Delta f_{\mathrm{FM}}= \frac{1}{2\pi} \frac{d\Theta(t)}{dt} = k_{f} \max \left[\left|M(t)\right|\right].
\end{aligned}
$$

A partir dos desvios de frequência $\Delta f$, podemos obter os índices de modulação $\beta$ associados a cada esquema de modulação de fase calculando

$$
\begin{equation}
\beta = \frac{\Delta f}{B}
\end{equation}
$$

em que $B$ é a largura de banda do sinal mensagem $M(t)$. Portanto, o índice de modulação representa um fator de multiplicidade entre a banda do sinal modulado e a banda do sinal modulante.

### Modulação de faixa-estreita

Se a modulação em ângulo for tal que $\Theta(t)\ll 1$, podemos aproximar $\cos\Theta(t)\approx 1$ e $\operatorname{sen}\Theta(t)\approx \Theta(t)$, de modo que

$$
\begin{align}
Y(t) &= A\cos\left[2\pi f_ct + \Theta(t)\right]\nonumber\\
     &= A\cos(2\pi f_ct)\cos\Theta(t)-A\operatorname{sen}(2\pi f_ct)\operatorname{sen}\Theta(t)\nonumber\\
     &\approx A\cos(2\pi f_ct) - A\operatorname{sen}(2\pi f_ct)\Theta(t)
\end{align}
$$


**Autocorrelação e densidade espectral de potência**

O sinal modulado $Y(t)$ é um processo cicloestacionário definido por com função de autocorrelação dada por

$$\begin{equation} 
R_{Y}(t+\tau, t) = E\left[Y(t)Y(t+\tau)\right] 
\end{equation}
$$

Podemos expandir o produto $Y(t)Y(t+\tau)$ na forma

$$
\begin{align} 
Y(t)Y(t+\tau) &= A^{2}\cos \left(2 \pi f_{c} t\right)\cos \left(2 \pi f_{c} t + 2 \pi f_{c} \tau\right)\nonumber\\
              &- A^{2}\operatorname{sen} \left(2 \pi f_{c} t\right)\cos \left(2 \pi f_{c} t + 2 \pi f_{c} \tau\right)\Theta(t)\nonumber\\
              &- A^{2}\cos \left(2 \pi f_{c} t\right)\operatorname{sen} \left(2 \pi f_{c} t + 2 \pi f_{c} \tau\right)\Theta(t+\tau)\nonumber\\
              &+ A^{2}\operatorname{sen} \left(2 \pi f_{c} t\right)\operatorname{sen} \left(2 \pi f_{c} t + 2 \pi f_{c} \tau\right)\Theta(t)\Theta(t+\tau)\nonumber\\
\end{align}$$

Assumindo que $M(t)$ é estacionário com média nula, temos que $E\left[M(t)\right]=E\left[\Theta(t)\right]=0$. Logo

$$
\begin{align} 
E\left[Y(t)Y(t+\tau)\right] &= A^{2}\cos \left(2 \pi f_{c} t\right)\cos \left(2 \pi f_{c} t + 2 \pi f_{c} \tau\right)\nonumber\\
              &- A^{2}\operatorname{sen} \left(2 \pi f_{c} t\right)\cos \left(2 \pi f_{c} t + 2 \pi f_{c} \tau\right)E\left[\Theta(t)\right]\nonumber\\
              &- A^{2}\cos \left(2 \pi f_{c} t\right)\operatorname{sen} \left(2 \pi f_{c} t + 2 \pi f_{c} \tau\right)E\left[\Theta(t+\tau)\right]\nonumber\\
              &+ A^{2}\operatorname{sen} \left(2 \pi f_{c} t\right)\operatorname{sen} \left(2 \pi f_{c} t + 2 \pi f_{c} \tau\right)E\left[\Theta(t)\Theta(t+\tau)\right]\nonumber\\
              &= \frac{A^{2}}{2}\left[\cos \left(2 \pi f_{c} \tau\right) + \cos \left(4 \pi f_{c} t + 2 \pi f_{c} \tau\right)\right] + \frac{A^{2}}{2}\left[\operatorname{cos} \left(2 \pi f_{c} \tau\right)-\cos \left(4 \pi f_{c} t + 2 \pi f_{c} \tau\right)\right]E\left[\Theta(t)\Theta(t+\tau)\right]
\end{align}
$$

Pelo corolário 2 do teorema de Wiener-Khinchin, temos

$$
\begin{align}
\bar{R}_{Y}(\tau) &= \frac{1}{T_{c}} \int_{-T_{c} / 2}^{T_{c} / 2} R_{Y}(t+\tau, t) dt\nonumber\\
                  &= \frac{A^2}{2}\cos(2\pi f_c \tau) + \frac{A^2}{2}R_{\Theta}(\tau)\cos(2\pi f_c \tau) 
\end{align}
$$

em que $T_c = 1/f_c$. 

Logo, a densidade espectral de potência $S_{Y}(f)$ do sinal modulado será dada por

$$
\begin{align}
S_{Y}(f) &= \mathscr{F}[\bar{R}_{Y}(\tau)]\nonumber\\
&= \frac{A^{2}}{2}\mathscr{F}\left[\cos\left(2 \pi f_{c} \tau\right)\right]\ast \mathscr{F}\left[1 + R_{\Theta}(\tau) \right] \nonumber\\
&= \frac{A^2}{2}\left(\frac{1}{2}\left[\delta(f-f_c)+\delta(f+f_c)\right]\ast \left[\delta(f) + S_{\Theta}(f)\right] \right)\nonumber\\
&= \frac{A^2}{4}\left[\delta(f-f_c)+\delta(f+f_c)\right] + \frac{A^2}{4}\left[S_{\Theta}(f-f_c)+S_{\Theta}(f+f_c)\right]\nonumber
\end{align}
$$

Portanto, no caso do sinal PM, temos $S_{\Theta}(f) =k_p^2S_M(f)$ e 

$$
\begin{equation}
S_{Y}(f) =  \frac{A^2}{4}\left[\delta(f-f_c)+\delta(f+f_c)\right] + \frac{A^2k_p^2}{4}\left[S_{M}(f-f_c)+S_{M}(f+f_c)\right].\nonumber
\end{equation}
$$

No caso do sinal FM, temos $S_{\Theta}(f) = k_f^2S_M(f)/f^2$ e 

$$
\begin{equation}
S_{Y}(f) =  \frac{A^2}{4}\left[\delta(f-f_c)+\delta(f+f_c)\right] + \frac{A^2k_f^2}{4}\left[\frac{S_{M}(f-f_c)}{(f-f_c)^2}+\frac{S_{M}(f+f_c)}{(f+f_c)^2}\right].\nonumber
\end{equation}
$$
